# Model Identification

```python
cars=data_new.copy()
```

**pd.get_dummies()**

```python
cars.info()
```

```python
dumies_list=list(cars.select_dtypes(include=['object']).columns)
```

```python
dumies_list
```

```python
#we converted categorical columns to numeric clones consisting of 1s and zeros
for i in dumies_list:
    temp_df=pd.get_dummies(eval('cars'+'.'+i),drop_first=True)
    cars=pd.concat([cars,temp_df],axis=1)
    cars.drop([i],axis=1,inplace=True)
```

## Train-Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
train_data, test_data = train_test_split(cars, train_size=0.8, random_state=1)
```

# Multiple linear regression

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
# let's get linear regression object
lr=LinearRegression()
```

```python
# train (fit) linear regression with data
lr.fit(X_train,y_train)
```

**RFE (Recursive Feature Elimination)**:

```python
rfe=RFE(lr,n_features_to_select=10)
```

```python
rfe=rfe.fit(X_train,y_train)
```

```python
rfe.support_
```

```python
rfe.ranking_
```

```python
X_train_rfe=X_train[X_train.columns[rfe.support_]]
```

```python
X_train_rfe
```

## OLS

```python
X_train_rfemodel=X_train_rfe.copy()
```

```python
X_train_rfemodel=sm.add_constant(X_train_rfemodel)
```

```python
lr=sm.OLS(y_train,X_train_rfemodel).fit()
```

```python
print(lr.summary())
```
